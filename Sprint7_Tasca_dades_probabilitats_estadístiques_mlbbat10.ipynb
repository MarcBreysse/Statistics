<span style=color:green> Mitjana aritmètica (Mean) amb pandas series std() method </span>

```python
Mlbbat10_mode_game = Mlbbat10['game'].mode()
round(Mlbbat10_mode_game, 1)
```

<span style=color:blue>Continuant amb les dades de tema esportiu, calcula la correlació de tots els atributs entre sí i representa'ls en una matriu amb diferents colors d'intensitat..</span>

```python
corr = Mlbbat10.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(Mlbbat10.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(Mlbbat10.columns)
ax.set_yticklabels(Mlbbat10.columns)
plt.show()
```

https://machinelearningmastery.com/arithmetic-geometric-and-harmonic-means-for-machine-learning/

<span style=color:blue>The central tendency summarizes the most likely value for a variable, and the average is the common name for the calculation of the mean.
The arithmetic mean is appropriate if the values have the same units, whereas the geometric mean is appropriate if the values have differing units.
The harmonic mean is appropriate if the data values are ratios of two variables with different measures, called rates.</span>


<span style=color:green> Mitjana geomètrica 'games' (gmean) </span>

```python
import scipy
from scipy import stats
Mlbbat10_gmean_game=scipy.stats.gmean(Mlbbat10.loc[:,'game'],axis=0)
Mlbbat10_gmean_game
```

```python
plt.hist(Mlbbat10['game'],edgecolor='k')
plt.title("number of games histogram",fontsize=12)
plt.xlabel("number of games")
plt.ylabel("Frecuency")
plt.show()
```

<span style=color:green> Mitjana geomètrica 'double' (gmean) </span>

```python
Mlbbat10_gmean_double=scipy.stats.gmean(Mlbbat10.loc[:,'double'],axis=0)
Mlbbat10_gmean_double
```

```python
plt.hist(Mlbbat10['double'],edgecolor='k')
plt.title("number of games histogram",fontsize=12)
plt.xlabel("number of games")
plt.ylabel("Frecuency")
plt.show()
```

<span style=color:green> Mitjana harmònica 'game' (hmean) </span>

```python
Mlbbat10_gmean_game=scipy.stats.hmean(Mlbbat10.loc[:,'game'],axis=0)
Mlbbat10_gmean_game
```
